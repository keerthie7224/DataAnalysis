
# Zomato Dataset - EDA Notebook

This notebook covers:
1. Import Libraries  
2. Load Dataset  
3. Explore Data  
4. Data Cleaning  
5. Analysis with NumPy & Pandas  
6. Visualization with Matplotlib  
7. Visualization with Seaborn


```python

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure display
pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python

# 2. Load Dataset
df = pd.read_csv("zomato.csv", encoding='latin-1')
print("Dataset Loaded Successfully!")
df.head()
```

```python

# 3. Explore Data
print("Shape of dataset:", df.shape)
print("\nColumn names:\n", df.columns.tolist())
print("\nBasic Info:")
df.info()
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary Statistics:")
df.describe(include='all').T
```

```python

# 4. Data Cleaning

# Drop duplicates
df.drop_duplicates(inplace=True)

# Clean 'Rate' column
if 'Rate' in df.columns:
    df['Rate'] = df['Rate'].astype(str).str.replace('/5', '').str.strip()
    df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
    df['Rate'].fillna(df['Rate'].median(), inplace=True)

# Fill missing values
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Convert 'Approx Cost(for two people)' to numeric
if 'Approx Cost(for two people)' in df.columns:
    df['Approx Cost(for two people)'] = df['Approx Cost(for two people)'].astype(str).str.replace(',', '').astype(float)

# Convert binary categorical columns
if 'Online Order' in df.columns:
    df['Online Order'] = df['Online Order'].map({'Yes': 1, 'No': 0})
if 'Book Table' in df.columns:
    df['Book Table'] = df['Book Table'].map({'Yes': 1, 'No': 0})

# Confirm cleaning
print("Missing values after cleaning:\n", df.isnull().sum())
df.head()
```

```python
print(df.columns.tolist())
```

```python

# 5. Analysis with NumPy & Pandas

# Average rating
avg_rating = np.mean(df['Aggregate rating'])
print("Average Rating:", round(avg_rating, 2))

# Average cost for two
if 'Approx Cost(for two people)' in df.columns:
    avg_cost = np.mean(df['Approx Cost(for two people)'])
    print("Average Cost for Two:", round(avg_cost, 2))

# Most common cuisines and location
top_cuisines = df['Cuisines'].value_counts().head(5)
top_locations = df['Locality'].value_counts().head(5)
print("\nTop 5 Cuisines:\n", top_cuisines)
print("\nTop 5 Locations:\n", top_locations)
```

```python

# 6. Visualization with Matplotlib

# Rating distribution
plt.figure(figsize=(8,5))
plt.hist(df['Aggregate rating'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()

# Cost vs Rating scatter

plt.scatter(df['Average Cost for two'], df['Aggregate rating'], color='red', alpha=0.5)
plt.title("Approx Cost vs Rating")
plt.xlabel("Approx Cost (for two people)")
plt.ylabel("Rating")
plt.show()

# Bar chart of top cuisines
plt.figure(figsize=(10,5))
plt.bar(top_cuisines.index, top_cuisines.values, color='green')
plt.title("Top 5 Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python

# 7. Visualization with Seaborn

# Online Order vs Rating
plt.figure(figsize=(6,4))
sns.boxplot(x='Has Online delivery', y='Aggregate rating', data=df)
plt.title("Online Order vs Rating")
plt.show()

# Book Table vs Rating
plt.figure(figsize=(6,4))
sns.boxplot(x='Has Table booking', y='Aggregate rating', data=df)
plt.title("Book Table vs Rating")
plt.show()

# Heatmap for numeric correlation
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=True)
plt.title("Correlation Heatmap of Numeric Columns")
plt.show()
```
